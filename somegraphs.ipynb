```python
import matplotlib.pyplot as plt
import torch
```

```python
list = torch.rand(10)

plt.plot(list)
```

```python
zeros = torch.zeros(10,10, dtype=float)
zeros[1,1] = 1
```

```python
plt.imshow(zeros)
```

```python
identity = torch.zeros(10,10)
```

```python
d_tups = [(i,j) for i,j in zip(range(10),range(10))]
for i,j in d_tups:
    identity[i,j] = 1
```

```python
plt.imshow(identity)
```

```python
identity.diag()
```

```python
N = torch.rand(10,10)

N @ identity

fig, axs = plt.subplots(1,2)

axs[0].imshow(N @ identity)
axs[1].imshow(N)
plt.show()
```
